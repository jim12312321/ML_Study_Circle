# Colab 簡介
---------
Colab為google開發，基於Jupyter notebook的免費虛擬機，可供使用者進行python編寫，更可用於資料分析、機器學習等專案


# Colab 功能簡介


## 筆記本
---
### 新增筆記本
  * 檔案$\rightarrow$新增筆記本

### 開啟筆記本
  * 檔案$\rightarrow$開啟筆記本

### 儲存筆記本
  * 檔案$\rightarrow$儲存
  * **Ctrl+S**

### 重新命名筆記本
  * 檔案$\rightarrow$重新命名
  * 單擊上方名稱區直接重新命名

### 下載筆記本
  * 檔案$\rightarrow$下載


## 儲存格
---------
可供寫程式或是紀錄文字的區塊


### 執行儲存格
  * **Shift+Enter**（執行當前儲存格）
  * 執行階段 $ \rightarrow $ 執行儲存格（可根據需求選擇執行的區段）

### 中斷或重啟儲存格
  * 執行階段$\rightarrow$中斷/重啟執行程序（可根據需求選擇執行的方式）

### 刪除儲存格
  * **Ctrl+M+D**
  * 點選儲存格右上方功能列中的垃圾桶圖示
  * 編輯 $ \rightarrow $ 刪除所選儲存格

### 移動儲存格
  * 點選儲存格右上方功能列中的上箭頭或下箭頭圖示進行上移或下移



### 程式碼儲存格
---
* 新增方法
> 可由「插入$ \rightarrow $程式碼儲存格」新增程式碼儲存格，或是將鼠標移至儲存格上下邊界，點選「+ 程式碼」

```python
print('這個是程式碼儲存格')
```

### 文字儲存格
---
- 新增方法
> 可由「插入$ \rightarrow $文字儲存格」新增程式碼儲存格，或是將鼠標移至儲存格上下邊界，點選「+ 文字」

- 編輯方式
> 於文字儲存格上用鼠標雙擊即可編輯

- 編輯格式
> 主要使用**Markdown**語法，詳情可參考 [markdown
guide](https://markdown.tw/)，
<br>若需要使用數學符號或是公式，也可用 [LaTex](https://zh.wikipedia.org/wiki/LaTeX) 語法，
<br>僅需於LaTex語法前後加上**`$`**符號，例如`$\sqrt{3x-1}+(1+x)^2$`轉換成$\sqrt{3x-1}+(1+x)^2$
