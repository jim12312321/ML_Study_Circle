## 資料型態_結構
---
### 結構型態
|型態|名稱|語法|可否更改內容|
|-|-|-|-|
|`list`|串列|`[]`|可|
|`tuple`|多元組|`()`|不可|
|`dict`|字典|`{}`|可|

### `list`
* 簡介
  - 有**順序**的保存資料
  - 使用 `:` 來取得指定位置範圍中的資料
* 建立list
  - 若要建立空串列，可直接用`[]`，例如：`a_list=[]`
  - 若要初始化串列值，需在`[]`中輸入初始值，例如：`a_list=[1,2,3]`
  - 若要模擬2維以上list，可於`[]`中再放入list，例如`a_list = [[1,2],[3,4]]`
* 新增資料
  - 利用`append()`在list的**最後面**加入資料
  - 利用`insert()`在list中**插入**資料
* 移除資料
  - 利用`pop()`移除指定**位置**的資料
  - 利用`remove()`移除指定**元素值**
  - 利用`del`移除**連續範圍**資料
  - 利用`clear()`移除list中**所有資料**

### `tuple`
* 簡介
  - 類似於 `list` ，有**順序**的保存資料
  - 其使用方式大致上與 `list` 相同，但**無法變更個別元素的值**
* 建立tuple
  - 若要初始化tuple值，需在`()`中輸入初始值，例如：`a_tuple=(1,2,3)`

### `dict`
* 簡介
  - **無**順序的保存資料
  - 使用任意**數值型態**作為**鑰匙（鍵，key）** 取得**配對值（value）**
  - 同一個key只能對應到一個value，但同一個value可以賦予給多個key
* 建立dict
  - 若要建立空字典，可直接用`{}`，例如：`a_dict={}`
  - 若要初始化串列值，需在`{}`中輸入初始值，格式為`key : value`，例如：`a_list=['a':1,'b':2,'c':3]`
* 新增資料
  - 如同宣告新變數，`dict[new_key]=new_value`
  - 若要新增多筆資料，可利用`update()`
* 移除資料
  - 利用`pop()`移除指定**key**的資料，若key不存在，可設預設返回值
  - 利用`del`移除指定**key**的資料，若key不存在則報錯

### 資料運算

- 可以使用 `type` 觀察變數**資料型態**
- 可以使用 `len()` 取得結構型態變數的**大小**

````python
# list

empty_list = []  #可建立空串列
a_list = [1,2,3,4,5] #也可直接初始化list中的值
print(a_list)
print(type(a_list)) #輸出<class 'list'>
print(len(a_list),end='\n\n')  #輸出list中所含元素個數

n2_list = [[1,2,3],[4,5,6]] #建立2維list
print(n2_list)
print(n2_list[0][0]) #輸出(0,0)的元素
print(n2_list[0][:2]) #輸出第0列，0-1欄的元素
print(n2_list[0][1:3]) #輸出第0列，1-2欄的元素
print(n2_list[0][:]) #輸出第0列所有的元素
print(type(n2_list)) #輸出<class 'list'>
print(len(n2_list),end='\n\n')  #輸出2維list中所含元素個數

a_list.append(7)  #在list最後面加入值
print(a_list)
a_list.insert(5,6)  #在位置(5)插入值(6)
print(a_list,end='\n\n')

print(a_list[0])  #輸出list中位置(0)的值
print(a_list[0:4],end='\n\n')  #輸出list中位置(0~3)的值

a_list.pop(0)   #移除位置(0)中的值
print(a_list)
a_list.remove(7)   #移除值(7)
print(a_list)
del a_list[2:4]   #移除位置(2~3)中的值
print(a_list)
a_list.clear()   #移除所有值
print(a_list)
````

````python
# tuple

a_tuple = (1,2,3)
print(a_tuple)
print(type(a_tuple)) #輸出<class 'tuple'>
print(len(a_tuple))  #輸出tuple中所含元素個數
print(a_tuple[0])     #輸出tuple中位置(0)的值
print(a_tuple[0:2])   #輸出tuple中位置(0~1)的值
````

````python
# dict

empty_dict = {}  #可建立空字典
a_dict = {'a':1,'b':1,'c':2,'d':4} #也可直接初始化字典中的值
print(a_dict)
print(type(a_dict)) #輸出<class 'dict'>
print(len(a_dict),end='\n\n')  #輸出dict中所含元素個數

a_dict['e'] = 5   #新增key和value
print(a_dict,end='\n\n')

print(a_dict['b'],end='\n\n')  #輸出list中key(b)的值

a_dict.pop('a')   #移除key(a)中的值
print(a_dict)
del a_dict['c']   #移除key(c)中的值
print(a_dict)
````

### 條件運算式 



### `if` 語句

- 在條件後面須加上`:`
- 使用**縮排**的數量（Indentation Level）來判斷`if`的範圍
- 相鄰的兩行程式碼若有一樣的縮排數的話，會被視為在同一個執行環境裡面
- 常用的縮排為：2 個空白鍵、4 個空白鍵、1 個 tab 等
- 注意在同一個執行環境內必須使用相同的縮排，否則會出現錯誤：Indentation Error
- 在 python 中最簡單的流程控制語句為 `if...else` 語句
    - `if` 語句僅在條件判斷式為 `True` 時執行 `if` 執行環境內的指令：
    ```python
    if condition:
        some_statement # 當 condition 為 True 時執行 some_statement
    ```
    - 若要串接多個條件判斷式，可使用 `if...elif` 語句：
    ```python
    if condition_1:
        some_statement_1 # 當 condition_1 為 True 時執行 some_statement_1
    elif condition_2:
        some_statement_2 # 當 condition_1 為 False
                         # 且 condition_2 為 True 時執行 some_statement_2
    elif condition_3:
        some_statement_3 # 當 condition_1 與 condition_2 皆為 False
                         # 且 condition_3 為 True 時執行 some_statement_3
    ```
    - 若列舉的條件以外有統一的處理方式，可使用 `if...else` 語句：
    ```python
    if condition:
        some_statement_1 # 當 condition 為 True 時執行 some_statement_1
    else:
        some_statement_2 # 當 condition 為 False 時執行 some_statement_2
    ```
- 如果 `if` 語句後面只有一個指令要做的話，可以簡單的寫成一行：
    ```python
    if condition: expression
    ```
- 如果 `if...else` 語句，`if` 和 `else` 的後面都只有一個指令要做的話，可以簡單寫成一行：
    ```python
    expression_1 if condition else expression_2
    ```

````python
# 條件運算子

print(1==1)       #輸出True
print(1!=2)       #輸出True
print(1<2)        #輸出True
print(1<=1)       #輸出True
print(2>1)        #輸出True
print(2>=1)       #輸出True
print(3 in [1,2,3])   #輸出True
````

````python
# 邏輯運算子

# and 
print(True and True)    #輸出True
print(True and False)   #輸出False
print(False and True)   #輸出False
print(False and False,end='\n\n')   #輸出False

# or 
print(True or True)    #輸出True
print(True or False)   #輸出True
print(False or True)   #輸出True
print(False or False,end='\n\n')   #輸出False

# not
print(not True)    #輸出False
print(not False)   #輸出True
````

````python
#條件判斷

Temperature = 30
if Temperature > 28: #只有if
  print('開冷氣')

if Temperature > 28 : print('開冷氣') #改寫成一行

score = 30
if score > 80: 
  print("A")
elif score > 60: #利用elif串接
  print("B")
else:         #加上else
  print("F")

print('pass') if score > 60 else print('fail') #若if...else型態中expression只有一行，可將整段改寫成一行
````

## 迴圈
---
### `for` 迴圈

- 與 `if` 語句相同皆須加`:`與縮排
    ```python
    for variable in iterable:
        some_statement # Do something
    ```
- `iterable` 可以為 `list`, `str`, `tuple`, `dict` 或是 `range` 函式
    - `list`, `str`, `tuple` 會列舉每個位址中的值
    - `dict` 會列舉所有的key
    - `range` 會按照需求生成值直到滿足停止條件
- `range` 函式
    - start 為起始值，end 為終止值(不包含)，step 為遞增(減)值 (非必要)
    ```python
    range(start)
    range(start, end)
    range(start, end, step)
    ```
    
### 巢狀迴圈  

- 若迴圈內容受到兩個 (或兩個以上) 的變數來分別控制其變化，此時可使用巢狀迴圈
    ```python
    for variable_1 in iterable_1:
        some_statement_1 # Do something
        for variable_2 in iterable_2:
            some_statement_2 # Do something
    ```

### `while` 迴圈

- 與 `if` `for` 語句相同皆須加`:`與縮排
    ```python
    while (condition):
        some_statement # Do something
    ```
- `while` 迴圈為先計算是否滿足條件，若滿足再執行迴圈內容

### `break`, `continue`  語句

- `break` 用途為中斷迴圈的執行並**跳脫迴圈**，繼續執行迴圈外的敘述
- `continue` 用途為跳過迴圈內 `continue` 後面的剩餘敘述，接著繼續執行下一次的迴圈運作，**不會跳脫迴圈**

````python
#for loop

for i in range(5): #利用range
  print(i,end=' ')

print('\n\n')

for i in ['a','b','c']: #利用list
  print(i,end=' ')

print('\n\n')

for i in 'zxc': #利用str
  print(i,end=' ')

print('\n\n')

for i in {'a':1,'b':2,'c':3}: #利用dict
  print(i,end=' ')
````

````python
#for 巢狀迴圈
for i in range(2):
  print('\n')
  print('i is {}'.format(i))
  print('---------')
  for j in range(3):
    print('j is {}'.format(j))
````

````python
#while loop
people = 3
while(people > 0):
  print('人數:{}'.format(people))
  people-=1
````

````python
#break 

for i in range(6):  
  if i == 3: break
  print('i is {}'.format(i))

print('\n\n')

#continue 

for i in range(6):  
  if i == 3: continue
  print('i is {}'.format(i))
````
